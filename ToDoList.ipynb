```python
import matplotlib.pyplot as plt
import networkx as nx


# Define edges to represent relationships (you can customize this based on your understanding of relationships)
edges = [
    ("Research", "Photogrammetry"),
    ("Research", "Publications"),
    ("Research", "Grant and Funding Applications"),
    ("Research", "Student Involvement"),
    ("Education, Outreach, and Public Engagement", "Workshops"),
    ("Workshops", "Photogrammetry"),
    ("Education, Outreach, and Public Engagement", "Public Tours"),
    ("Education, Outreach, and Public Engagement", "Media Content"),
    ("Education, Outreach, and Public Engagement", "Grant and Funding Applications"),
    ("Education, Outreach, and Public Engagement", "Student Involvement"),
    ("Data Archival and Preservation", "Education, Outreach, and Public Engagement"),
    ("LPL Support", "Class Support"),
    ("LPL Support", "Grant and Funding Applications")
]

# Use set comprehensions to get unique strings
unique_strings = {string for edge in edges for string in edge}
functions = list(unique_strings)

importance_values = [1, 4, 4, 3, 2, 4, 3, 4, 4, 5, 4, 5]

# Create a directed graph
G = nx.DiGraph()

# Add nodes with labels and sizes
for function, importance in zip(functions, importance_values):
    G.add_node(function, size=importance)

# Add edges to the graph
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(18, 12))
pos = nx.spring_layout(G)  # You can use different layout algorithms
node_sizes = [importance * 100 for importance in importance_values]  # Scale for better visualization
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='skyblue', alpha=0.8)
nx.draw_networkx_labels(G, pos, font_size=10, font_color="black", font_weight="bold",)
nx.draw_networkx_edges(G, pos, edge_color='gray', alpha=0.5, arrows=True)

plt.title('Directed Graph of Space Imagery Center Functions and Relationships')
plt.axis('off')
plt.show()
```

# Space Imagery Center Redesign Action Items

```python
# Import necessary libraries
import networkx as nx
import matplotlib.pyplot as plt

# Function to create a directed graph
def create_dependency_graph(dependencies):
    G = nx.DiGraph()

    # Add nodes and edges based on dependencies
    for item, dependent_items in dependencies.items():
        G.add_node(item)
        G.add_edges_from((item, dependent_item) for dependent_item in dependent_items)

    return G

# Define your action items and their dependencies
item_a = """Find storage for items 
on bookshelves 
and in cabinets"""
item_b = """Remove bookshelves 
and cabinets from SIC"""
item_c = """Locate meteorites and other rock 
samples for display (Drake building)"""
item_d = """Locate or purchase spacecraft 
models for display"""
item_e = 'Purchase display cases'
item_f = """Purchase topo table 
from White Clouds"""
item_g = 'Purchase AR Sandbox'
item_h = 'Set up new display cases'

action_items = {
    item_a: [],
    item_b: [item_a],
    item_c: [],
    item_d: [],
    item_e: [],
    item_f: [item_b],
    item_g: [item_b],
    item_h: [item_b, item_c, item_d, item_e]
}

# Display action items as a bullet-point list
print("### Action Items ###")
for item, dependent_items in action_items.items():
    print(f"- {item}")
    if dependent_items:
        print(f"  - Depends on: {', '.join(dependent_items)}")

# Create a directed graph to represent dependencies
dependency_graph = create_dependency_graph(action_items)

# Plot the graph with improved layout
plt.figure(figsize=(12, 8))
pos = nx.circular_layout(dependency_graph)
# pos = nx.spring_layout(dependency_graph, seed=6)  # Use seed for reproducibility
nx.draw(dependency_graph, pos, with_labels=True, node_size=1000, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", arrowsize=20, edge_color='gray', linewidths=0.5)

plt.title("Action Item Dependencies")
plt.show()
```
